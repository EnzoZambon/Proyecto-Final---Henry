```python
df['season_id'] = df['season_id'].astype(str).str[1:].astype(int)
```

```python
columnas_a_eliminar = [
    'video_available_away', 
    'video_available_home',"min"]
df = df.drop(columns=columnas_a_eliminar)
```

```python
missing_data = df.isnull().mean() * 100
print(missing_data[missing_data > 0].sort_values(ascending=False))
```

```python
columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns
columnas_a_llenar = [col for col in columnas_numericas if col in [
    'fgm_home', 'fga_home', 'fg_pct_home', 'fg3m_home', 'fg3a_home', 'fg3_pct_home', 
    'ftm_home', 'fta_home', 'ft_pct_home', 'oreb_home', 'dreb_home', 'reb_home', 
    'ast_home', 'stl_home', 'blk_home', 'tov_home', 'pf_home', 'pts_home', 
    'plus_minus_home', 'fgm_away', 'fga_away', 'fg_pct_away', 'fg3m_away', 'fg3a_away', 
    'fg3_pct_away', 'ftm_away', 'fta_away', 'ft_pct_away', 'oreb_away', 'dreb_away', 
    'reb_away', 'ast_away', 'stl_away', 'blk_away', 'tov_away', 'pf_away', 'pts_away', 
    'plus_minus_away'
]]


for columna in columnas_a_llenar:
    df[columna] = df[columna].fillna(0)
```

```python
missingData = df.isnull().mean() * 100
print(missingData[missingData > 0].sort_values(ascending=False))
```

```python
df['wl_home'] = df['wl_home'].fillna('unknown')
df['wl_away'] = df['wl_away'].fillna('unknown')
```

```python
df = df.drop_duplicates(subset='game_id', keep='first')
```

```python
columns_to_keep = ['game_id', 'season_id', 'game_date', 'team_id_home', 'team_id_away', 'pts_home', 'pts_away', 'season_type']
filtered_games = df[columns_to_keep]

filtered_games = filtered_games.rename(columns={
    'team_id_home': 'home_team_id',
    'team_id_away': 'away_team_id',
    'pts_home': 'home_points',
    'pts_away': 'away_points'
})
```
