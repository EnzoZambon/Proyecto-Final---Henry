```python
df = df[df['eventmsgtype'] != 18]
```

```python
df['player1_id'] = df['player1_id'].fillna(0).astype('Int64')
df['player2_id'] = df['player2_id'].fillna(0).astype('Int64')
df['player3_id'] = df['player3_id'].fillna(0).astype('Int64')
```

```python
repeticiones = df[['game_id', 'eventnum']].duplicated(keep=False)

# Filtrar las filas duplicadas
filas_duplicadas = df[repeticiones]

# Mostrar las filas duplicadas
print(filas_duplicadas)
```

```python
df = df.drop_duplicates(subset=['game_id', 'eventnum'])
```
