```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Enzo/OneDrive/Escritorio/Henry PF/csv/other_stats.csv")
df.head(5)
```

```python
missingData = df.isnull().mean() * 100
print(missingData[missingData > 0].sort_values(ascending=False))
```

```python
valoresFaltantesColumnas = df.columns[df.isnull().any()]

# Imprimir las columnas con valores nulos
print("Columnas con valores nulos:\n", valoresFaltantesColumnas)

# Rellenar valores nulos con el promedio de la columna agrupado por equipo
for column in valoresFaltantesColumnas:
    df.loc[:, column] = df.groupby("game_id")[column].transform(lambda x: x.fillna(0))

# Mostrar la tabla resultante
print("Tabla después de la imputación de valores nulos:\n", df)
```
